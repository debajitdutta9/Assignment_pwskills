Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of
Matplotlib.

Matplotlib is a Python library used for creating data visualizations. It is one of the most widely used libraries for creating static, interactive, and animated visualizations in Python. Matplotlib provides a range of customizable 2D and 3D plots and graphics that can be used for data analysis, scientific research, and other applications.

Matplotlib is used for various purposes, including:

Data visualization: Matplotlib can be used to create different types of charts and plots such as line plots, scatter plots, bar plots, histogram plots, and more.

Exploratory data analysis: Matplotlib provides tools for visualizing and analyzing data to discover patterns, trends, and relationships.

Scientific research: Matplotlib is widely used in scientific research for data visualization and analysis.

Presentation graphics: Matplotlib can be used to create high-quality graphics for presentations and publications.

Web application development: Matplotlib can be used to create interactive visualizations for web applications.

Five plots that can be plotted using the Pyplot module of Matplotlib are:

Line plot: A line plot is a type of chart that displays data as a series of points connected by straight lines.

Scatter plot: A scatter plot is a type of chart that displays data as a collection of points. Each point represents an observation in the data set.

Bar plot: A bar plot is a type of chart that displays data as rectangular bars. The height or length of the bars represents the values in the data set.

Histogram plot: A histogram plot is a type of chart that displays the distribution of a numerical variable. It groups the data into intervals or bins and displays the frequency or count of observations in each bin.

Pie chart: A pie chart is a type of chart that displays data as a circle divided into slices. The size of each slice represents the proportion of the data in a particular category or group.

Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.
import numpy as np

Note: Also add title, xlabel, and ylabel to the plot.

scatter plot :A scatter plot is a type of chart that displays data as a collection of points. Each point represents an observation in the data set, and the position of the point on the x and y-axis represents the values of two variables.

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.scatter(x,y)
plt.title('Scatter graph')
plt.xlabel('x axis')
plt.ylabel('y axis')
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

A bar plot is a graphical representation of data using rectangular bars of equal width. The height of each bar represents the value of a particular variable or data point. Bar plots are commonly used to show comparisons between different categories or groups of data.

Bar plots are useful because they allow you to quickly and easily compare different values or categories. They can also be used to display data that has been sorted in ascending or descending order. Bar plots are commonly used in many different fields, including finance, economics, marketing, and science, to visually represent data and help interpret and communicate findings to others.

Overall, bar plots are a simple yet effective way to display and compare data, making them a useful tool in data analysis and communication.

```python
import numpy as np
import matplotlib.pyplot as plt
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.bar(company,profit)
plt.title('this is bar graph')
plt.xlabel('company')
plt.ylabel('profit')
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

A box plot, also known as a box-and-whisker plot, is a graphical representation of data that shows the distribution of a dataset. It is commonly used to display the range, median, quartiles, and outliers of a dataset.

A box plot consists of a rectangular box that spans the interquartile range (IQR), which is the distance between the first and third quartiles. Inside the box, a line represents the median of the data. Lines, called whiskers, extend from the box to the smallest and largest values that are not considered outliers. Outliers are plotted individually as points outside of the whiskers.

Box plots are useful because they provide a compact summary of the distribution of the data. They allow you to easily compare multiple datasets or variables and identify any outliers or extreme values. Box plots are commonly used in many different fields, including statistics, data science, and machine learning, to explore and analyze data.

Overall, box plots are a useful tool for visualizing and summarizing data. They provide a quick and easy way to see the distribution of a dataset and help identify any unusual or extreme values.

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate data
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Create a list of data to plot
data_to_plot = [box1, box2]

# Create a figure and axis object
fig, ax = plt.subplots()

# Create the box plot
bp = ax.boxplot(data_to_plot)

# Add labels and title
ax.set_xticklabels(['Box 1', 'Box 2'])
ax.set_ylabel('Value')
ax.set_title('Box plot of Box 1 and Box 2')

# Display the plot
plt.show()
```
